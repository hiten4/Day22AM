```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.DataFrame({
 'category': np.random.choice(['A','B','C'], 500),
 'value': np.random.randint(10,100,500),
 'value2': np.random.randint(5,50,500)
})
```

```python
grouped = df.groupby('category')['value'].mean()
grouped
```

```python
grouped.plot(kind='bar')
plt.show()
```

```python
df['value'].plot(kind='kde')
df['value2'].plot(kind='kde')
plt.show()
```
